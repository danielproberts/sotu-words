```python
## Include Statements

import json
import requests
import ast
import matplotlib.pyplot as plt
import pandas as pd
import sqlite3
import os
import re
```

```python
# Build CSV File

f = open("sotuwords.csv", "w+")
f.write("name,date,countWar,countPeace,countGod,countProvidence,countHeaven\n")
    
for filename in os.listdir(os.getcwd() + "/state-of-the-union-address-texts"):
    name = filename.split("[")
    name = name[0]
    name = name[:-1]
    date = filename[-10:-5]
    
    filepath = os.getcwd() + "/state-of-the-union-address-texts/" + filename
    
    reWar = "[\s,.]war[\s,.'s][\ss]?"
    rePeace = "[\s,.]peace[\s,.'ful']?"
    reGod = "[\s,.]god[\s,.'s][\ss]?" #[\s,.''s']"
    reProvidence = "[\s,.]providence[\s,.]"
    reHeaven = "[\s,.]heaven[\s,.]"
    #reTerrorism = "[\s,.]terrorism[\s,.]"
    #reTerrorist = "[\s,.]terrorist[\s,.]"

    file  = open(filepath, 'r', encoding = "ISO-8859-1").read()
    countWar = len(re.findall(reWar, file, re.IGNORECASE))
    countPeace = len(re.findall(rePeace, file, re.IGNORECASE))
    countGod = len(re.findall(reGod, file, re.IGNORECASE))
    countProvidence = len(re.findall(reProvidence, file, re.IGNORECASE))
    countHeaven = len(re.findall(reHeaven, file, re.IGNORECASE))
    #countTerrorism = len(re.findall(reTerrorism, file, re.IGNORECASE))
    #countTerrorist = len(re.findall(reTerrorist, file, re.IGNORECASE))
    
    #count = file.count()
    
    f.write(name + ",")
    f.write(date + ",")
    f.write(str(countWar) + ",")
    f.write(str(countPeace) + ",")
    f.write(str(countGod) + ",")
    f.write(str(countProvidence) + ",")
    f.write(str(countHeaven) + "\n")
    #f.write(str(countTerrorism) + "\n")
    #f.write(str(countTerrorist) + "\n")
    #f.write(str(count) + "\n")
```

```python
## Create Dataframe from File

data = pd.read_csv("sotuwords.csv", index_col=1)
data = data.sort_index()
print(data)
```

```python
## Create Database from Data Frame

conn = sqlite3.connect("sotuwords.db")
data.to_sql("sotuwords", conn, if_exists="replace")
```

```python
# SQL command

sql = "SELECT date, name, countWar, countPeace FROM sotuwords WHERE date > 1947"

df = pd.read_sql(sql, conn)
```

```python
print(df)
```

```python
plt.figure(figsize=(20,10))
plt.rcParams.update({'font.size': 22})
plt.plot(df.date, df.countWar, label="War", color="red")
plt.plot(df.date, df.countPeace, label="Peace", color="green")
plt.xlabel('Year')
plt.ylabel('Word Count')
plt.legend(loc='upper right')
plt.show()
```
